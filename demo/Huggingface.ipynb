# **Pipeline**

## 文本分类任务
model, tokenizer: 这个任务默认的模型

```python
from transformers import pipeline

pipe=pipeline("text-classification")
input="transformer is a useful tool for me, I use it to write program"
res=pipe(input)

print(res)
```

##  命名实体识别（Named Entity Recognition, NER
它的主要功能包括识别文本中的命名实体类别，比如：
人名（PERSON）
地名（LOCATION）
组织（ORGANIZATION）
日期/时间（MISC，Miscellaneous）
这个模型适合处理英语文本，能够自动标注输入句子中的实体类别。

# transformers
transformers 是 Hugging Face 提供的一个功能强大的 Python 库，专门用于加载、使用和微调 预训练的深度学习模型，特别是基于 Transformer 架构的模型。它是 NLP 和 CV 等领域的主流工具之一，支持各种任务如文本生成、分类、翻译和问答。

```python
from transformers import AutoModelForTokenClassification,AutoTokenizer

model=AutoModelForTokenClassification.from_pretrained("dbmdz/bert-large-cased-finetuned-conll03-english")
tokenizer=AutoTokenizer.from_pretrained("google-bert/bert-base-cased")
pipe=pipeline("ner",model=model,tokenizer=tokenizer)

res1=pipe("hello, my name is Jiang Fang")
res2=pipe("Li Lie is my best friend")
```

```python
print(res1)
print(res2)
```

## **Tokenizer**
分词器

```python
str="transformer is a useful tool for me, I use it to write program"

# return BatchEncodeing, with field input_ids, attention_mask......
tokens=tokenizer(str)
print(tokens)
print(tokens.input_ids)

# return pytorch tensor
tokens_pt=tokenizer(str,return_tensors="pt")
print(tokens_pt)

# tokenizer的其他方法
tokens2ids=tokenizer.convert_tokens_to_ids(str)
print(tokens2ids)
print(tokenizer.tokenize(str))
print(tokenizer.decode(tokens2ids))
```

## **model+tokenizer**

```python
from transformers import AutoTokenizer,AutoModel
import torch

task="text-classification"
model_name="distilbert/distilbert-base-uncased-finetuned-sst-2-english"
x_train=["hello, my name is lili","glad to see you"]


model=AutoModel.from_pretrained(model_name)
tokenizer=AutoTokenizer.from_pretrained(model_name)

input=tokenizer(x_train,return_tensors="pt")

with torch.no_grad():
  output=model(**input)
  print(output)
  logits=output.logits
  print(logits)
  res=logit.argmax(dim=-1)
  print(res)

```
